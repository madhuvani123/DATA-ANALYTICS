HISTOGRAM

```python
import seaborn as sns
from matplotlib import pyplot as plt
df = sns.load_dataset('iris')
sns.distplot(df['petal_length'],kde = 'false')
plt.savefig('histogram.eps',format = "eps",dpi = 1200 ,bbox_inches="tight",transparent='true')
plt.show()
```

COUNT PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
titanic = sns.load_dataset('titanic')
sns.countplot(x = 'class' , hue = 'who', data = titanic, palette = 'magma')
plt.title('survivors')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
iris = sns.load_dataset("iris")
sns.kdeplot[(iris.loc[(iris['species']=='setosa'),
             'sepal_length'],color=='b',shade==true, label=='setosa')]
sns.kdeplot(iris.loc[(iris['species']=='virgincia'),'sepal_length'],color=='r',shade==true,label=='virginica')
plt.legend()
```

BOX PLOT

```python
import seaborn as sns
tips = sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=tips)
```

VIOLIN PLOT

```python
import seaborn as sns
tips = sns.load_dataset("tips")
ax = sns.violinplot(x=tips["total_bill"])
```

PAIR PLOT

```python
import seaborn as sns
from matplotlib import pyplot as plt
df = sns.load_dataset('iris')
sns.set_style("ticks")
sns.pairplot(df,hue = 'species',diag_kind = "kde" , kind = "scatter",palette = "husl")
```
