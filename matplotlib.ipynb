```python
x= np.linspace(0,1,10)
y= 2*x
plt.grid(True)
plt.plot(x,y,'o:K')
plt.show()
```

```python
x= np.linspace(0,1,10)
y= 2*x
plt.figure(figsize=(10, 7))
plt.plot(x,y,':k')
plt.show()
```

```python
x = np.linspace(0,10,1000)
y=np.sin(x)
plt.plot(x,y,'o:r')
plt.show()
```

```python
x= np.linspace(0,10,1000)
y= np.sin(x)
z= np.cos(x)
plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

```python
x = np.linspace(0,10,100)
y= np.sin(x)
z= np.cos(x)
plt.figure(figsize=(20,10))
plt.plot(x,y, '*-r')
plt.plot(x,z, 'o-b')
plt.legend(("SINUSIOD","COSINE"))
plt.grid(True)
plt.show()
```

```python
x= np.linspace(0,10,100)
y= np.sin(x)
z= np.cos(x)
plt.figure(figsize=(20,10))
plt.plot(x, y, color= 'red', linestyle= 'solid', marker = '*')
plt.plot(x, z, color= 'blue', linestyle= 'solid', marker='o')
plt.legend(("SINUSIOD","COSINE"))
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
plt.figure(figsize=(20,10))
plt.plot(x, y , '*-r',linewidth=5)
plt.plot(x, z, 'o-b')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))

plt.subplot(2,2,3)
plt.plot(x , np.exp(x))

plt.subplot(2,2,4)
plt.plot(x , np.tan(x))

plt.show()
```

```python
t = np.arange(0.0 , 5.0 , 0.01)
s = np.cos(2*np.pi*t)
plt.plot(t , s , lw=2)
plt.annotate('important',xy=(2,1),xytext=(3,1.5),arrowprops=dict(facecolor='black'))
plt.ylim(-2,2)
plt.show()
```
